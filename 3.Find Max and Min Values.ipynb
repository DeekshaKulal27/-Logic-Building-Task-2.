```python
# Problem 3: Find Maximum and Minimum Values

numbers = [45, 22, 89, 10, 66]
maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Print results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```
